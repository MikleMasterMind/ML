```python
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt #     

# your code here
video_counts = df.groupby(df['publishedAt'].dt.date)['video_id'].nunique()
plt.figure(figsize=(10, 6))
plt.plot(video_counts.index, video_counts.values)
plt.xlabel('Publication Date')
plt.xticks(rotation=45)
plt.ylabel('Number of Released Video')
plt.title('Video Releases Over Time')
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
# your code here
video_counts_by_country = df.groupby('country')['video_id'].nunique()
plt.figure(figsize=(10, 6))
plt.bar(video_counts_by_country.index, video_counts_by_country.values)
plt.xlabel('Country')
plt.ylabel('Number of Published Video')
plt.title('Video Publications by Country')
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
# your code here
fr_rollers = df[(df['country'] == 'FR') & (df['comments_disabled'] == False)]
plt.figure(figsize=(10, 6))
plt.scatter(fr_rollers['view_count'], fr_rollers['comment_count'])
plt.xlabel('Number of Views')
plt.ylabel('Number of Comments')
plt.title('Comments vs Views for Rollers in France')
plt.show()
```
