```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
from sklearn.metrics import accuracy_score

parameters = {
    'n_neighbors': [i for i in range(1, 11)],
    'metrics': ['cosine'],
    'weights': ['uniform', 'distance'],
    'normalizers': [(count_vec, 'CountVectorizer'), (tf_idf, 'TfidfVectorizer')]
}
```

    ,    -    7 .   ,       colab .

```python
# 
from cross_val import kfold_split, knn_cv_score
from sklearn.neighbors import KNeighborsClassifier

folds = kfold_split(num_objects=len(X_train) , num_folds=3)
scores = knn_cv_score(X_train, y_train, parameters=parameters, score_function=accuracy_score, folds=folds, knn_class=KNeighborsClassifier)
```

```python
# 
count_vectorized_mean = np.mean([val for key, val in scores.items() if key[0] == 'CountVectorizer'])
tfidf_vectorized_mean = np.mean([val for key, val in scores.items() if key[0] == 'TfidfVectorizer'])

if count_vectorized_mean > tfidf_vectorized_mean:
    print('CountVectorizer')
elif count_vectorized_mean == tfidf_vectorized_mean:
    print('CountVectorizer', 'TfidfVectorizer')
else:
    print('TfidfVectorizer')
```

```python
# 
print(fetch_20newsgroups().target_names)

neighbors_score = dict()
for key, value in scores.items():
    n_neighbors = key[1]
    if n_neighbors in neighbors_score:
        neighbors_score[n_neighbors].append(value)
    else:
        neighbors_score[n_neighbors] = [value]
scores_mean = {n_neighbors: np.mean(scores) for n_neighbors, scores in neighbors_score.items()}
n_neighbors = list(scores_mean.keys())
mean = list(scores_mean.values())

plt.figure(figsize=(10, 10))
plt.grid()
plt.title('ependence of the quality metric on the number of neighbors')
plt.xlabel('n neighbors')
plt.ylabel('mean of the quality metric')
plt.plot(n_neighbors, mean)
plt.show()
```
